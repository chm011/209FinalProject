```python
import pandas as pd
import json
import math
```

```python
file_path = 'diabetes_binary_5050split_health_indicators_BRFSS2015.csv'
df = pd.read_csv(file_path)

df.head()

```

```python
def calculate_prevalence(group):
    print(f"Group Name: {group.name}")  # Print the group identifier
    print(group.head())  # Print the first few rows of the group
    total = group.count()
    diabetic_count = group.sum()
    print(f"Total: {total}, Diabetic Count: {diabetic_count}")
    if total == 0:
        return 0  
    else:
        return (diabetic_count / total) * 100
```

```python
df['Age'].head()
```

```python
mapping = {
    1.0: '18-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54',
    5.0: '55-64', 6.0: '65-74', 7.0: '75-84', 8.0: '85+',
    9.0: '90+', 10.0: '95+', 11.0: '100+', 12.0: '105+', 13.0: '110+'
}
df['AgeGroup'] = df['Age'].map(mapping)
print(df[['Age', 'AgeGroup']].head())


age_prevalence = df.groupby('AgeGroup')['Diabetes_binary'].mean().multiply(100).to_dict()

# BMI Grouping
bmi_bins = [0, 18.5, 25, 30, 100]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMICategory'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels, right=False)
bmi_prevalence = df.groupby('BMICategory')['Diabetes_binary'].mean().multiply(100).to_dict()

age_prevalence


```

```python
sex_prevalence = df.groupby('Sex')['Diabetes_binary'].mean().multiply(100).to_dict()
smoker_prevalence = df.groupby('Smoker')['Diabetes_binary'].mean().multiply(100).to_dict()
high_bp_prevalence = (df.groupby('HighBP')['Diabetes_binary'].mean().multiply(100).to_dict())
high_chol_prevalence = df.groupby('HighChol')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
heart_disease_prevalence = df.groupby('HeartDiseaseorAttack')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
phys_activity_prevalence = df.groupby('PhysActivity')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
gen_health_prevalence = df.groupby('GenHlth')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
ment_health_prevalence = df.groupby('MentHlth')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
phys_health_prevalence = df.groupby('PhysHlth')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
diff_walk_prevalence = df.groupby('DiffWalk')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
fruits_prevalence = df.groupby('Fruits')['Diabetes_binary'].apply(calculate_prevalence).to_dict()
```

```python
prevalence_data = {
    "Age": age_prevalence,
    "BMI": bmi_prevalence,
    "Sex": sex_prevalence,
    "Smoker": smoker_prevalence,
    "HighBP": high_bp_prevalence,
    "HighChol": high_chol_prevalence,
    "HeartDiseaseorAttack": heart_disease_prevalence,
    "PhysActivity": phys_activity_prevalence,
    "GenHlth": gen_health_prevalence,
    "MentHlth": ment_health_prevalence,
    "PhysHlth": phys_health_prevalence,
    "DiffWalk": diff_walk_prevalence,
    "Fruits": fruits_prevalence
}

prevalence_data
```

```python
with open('prevalence_data.json', 'w') as f:
    json.dump(prevalence_data, f)
```
